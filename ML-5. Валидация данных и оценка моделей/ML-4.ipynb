Регрессия - это класс задач обучения с учителем, когда по определённому набору признаков объекта необходимо предсказать числовую целевую переменную.

Выберите метрики, которые применяются в задаче классификации:

* Accuracy
* Precision
* Recall
* F1-score

Метрика Percision показывает, как много объектов, предсказанных моделью как положительные, на самом деле являются положительными.

Метрика Recall показывает, как много положительных объектов модель нашла среди всех положительных объектов.

Метрика R^2 score показывает, какую долю разнообразия (дисперсии) смогла уловить модель в данных.
