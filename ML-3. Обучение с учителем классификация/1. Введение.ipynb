

В общем случае уравнение линейной регрессии описывает уравнение плоскости в многомерном пространстве (гиперплоскости).

Высокий разброс модели — это вариативность ошибки на различных наборах данных. Это и есть переобучение: модель подстраивается под обучающий набор, уменьшая ошибку вплоть до 0, однако на других наборах ошибка модели может значительно отличаться.

Регуляризация — это метод борьбы с переобучением. Её суть заключается в намеренном увеличении смещения путём добавления в функцию потерь дополнительного слагаемого, именуемого штрафом. Благодаря этому ищется не истинный минимум функции потерь, а псевдоминимум.
