Перед нами пример неоднородной СЛАУ — системы, у которой хотя бы один коэффициент справа отличен от 0.



Если все коэффициенты справа равны 0, то система называется однородной СЛАУ:



Системы со множеством возможных решений называются неопределёнными и представляют особый интерес в линейной алгебре.



На этом всё. Мы нашли решение однородной системы, а точнее, решения. Вместо переменной  можно поставить любое число, и оно будет удовлетворять изначальной системе уравнений. Давайте это проверим.

Как называется матрица, составленная из коэффициентов при неизвестных переменных в СЛАУ?

Матрица системы

ЛИНЕЙНАЯ ЗАВИСИМОСТЬ СИСТЕМЫ ВЕКТОРОВ

Линейно зависимые векторы:




РАНГ МАТРИЦЫ. БАЗИС

```python
import numpy as np
v1 = np.array([1,2,3,1])
v2 = np.array([4,5,6,1])
v3 = np.array([7,8,9,1])
A = np.array([v1, v2, v3]).T
print(np.linalg.matrix_rank(A))
# 2
```

Мы видим, что ранг системы равен 2, а векторов три — значит система является линейно зависимой. В качестве базиса можем взять любые два вектора. Очевидно, каждая пара здесь является линейно независимой, т. к. никакие два вектора не пропорциональны.

```python
import numpy as np
v1 = np.array([1,2,3,1])
v2 = np.array([4,5,6,1])
v3 = np.array([7,8,10,1])
A = np.array([v1, v2, v3]).T
print(np.linalg.matrix_rank(A))
# 3
```

В данном примере ранг равен 3 — значит векторы линейно независимы и в базис войдут все три вектора

СВОЙСТВА РАНГА МАТРИЦЫ

Коллинеарные векторы всегда линейно зависимы по определению: один из векторов всегда можно умножить на какой-то коэффициент и получить второй.

Коллинеарные векторы не являются ортогональными, так как они сонаправлены друг другу, и угол между ними всегда 0 или 180 градусов, но никак не 90.
