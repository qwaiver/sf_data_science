Неверно:Разведывательный анализ данных (EDA) — это процесс поиска закономерностей в данных.

Неверно:Оптимизация гиперпараметров — это процесс подбора внешних параметров модели, а не внутренних.

Процесс поиска внутренних параметров модели называется обучением модели

Способ построения ансамбля путём последовательного объединения моделей таким образом, чтобы каждая новая модель уменьшала ошибку предыдущей композиции, называется бустингом

Дерево решений — это ациклический связный граф, вершинами которого являются условия (предикаты), на основе которых принимается решение о классе объекта.

* А Верно. GridSearchCV — это класс, реализующий метод перебора гиперпараметров на сетке.
* B Неверно. Pipeline — это класс пайплайна (конвеера), в котором описываются все этапы работы с данными от предобработки до построения модели.
* C Неверно. cross_validate — это функция, выполняющая процедуру кросс-валидации.
* D Верно. RandomSearchCV — это класс, реализующий метод перебора гиперпараметров случайным образом.

Линейная алгебра — это раздел алгебры, который изучает объекты линейной природы, векторы и векторные пространства, матрицы и линейные операторы, которые они порождают, и системы линейных уравнений.

Линейная алгебра лежит в основе решения всех основных задач Data Science: создания и настройки моделей, создания и обучения нейросетей, корреляционного анализа и т. д. В конце концов, даже табличный способ представления данных — это тоже фундаментальная основа линейной алгебры, матрица.

ЛИНЕЙНЫЕ МОДЕЛИ

Все линейные модели, такие как линейная и логистическая регрессия, берут своё начало из линейной алгебры. Сюда же относят метод опорных векторов (SVM).

МЕТОДЫ ПОНИЖЕНИЯ РАЗМЕРНОСТИ

В основе метода главных компонент (PCA), который мы неоднократно использовали для понижения размерности пространства признаков, лежит теория векторных пространств. Сингулярное разложение (SVD), о котором мы поговорим в модуле по рекомендательным системам, — это способ разложения матрицы на компоненты.

КОРРЕЛЯЦИОННАЯ МАТРИЦА

Напомним, что корреляция — это мера линейной зависимости между признаками. Если признаков несколько, то корреляция превращается в знакомую нам корреляционную матрицу. Матрицы и операции с ними изучает именно линейная алгебра.

ОБРАБОТКА ИЗОБРАЖЕНИЙ

Известно, что современные компьютеры работают только с числами 0 и 1. Тогда как в компьютере можно сохранить изображение, состоящее из различных цветов? Легко! Для этого изображение представляется в виде матрицы, отражающей интенсивность пикселей. Затем эта матрица может быть обработана для определения цветов, предметов и т. д.

Любая операция, которую вы хотите выполнить с этим изображением, вероятно, будет «под капотом» использовать линейную алгебру и матрицы.

Это лишь несколько очевидных примеров того, где линейная алгебра применяется в Data Science. На самом деле их намного больше, и мы поймём это, когда будем изучать, как методы машинного обучения устроены внутри.
