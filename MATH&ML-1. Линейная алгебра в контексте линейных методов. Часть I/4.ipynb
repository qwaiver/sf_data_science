Действия над матрицами условно можно разделить на базовые и продвинутые.

К базовым относятся следующие операции:

* Сложен*ие матриц. Результат — матрица.
* Умножение матрицы на число. Результат —  матрица.
* Транспонирование. Результат —  матрица.

Операция транспонирования — это матричная операция, при которой строки и столбцы меняются местами.



Обратите внимание, что при транспонировании главная диагональ матрицы не изменяется.

симметричные матрицы инварианты к операции транспонирования, то есть не изменяются.



B Неверно. Если матрица была прямоугольной, то при транспонировании она ей же и останется. Только из «высоких» прямоугольных матриц получаются «длинные» и наоборот.

C Неверно. Складывать можно только матрицы одинакового размера. В матрицах может быть равное количество элементов, но разные размерности: например, матрицы 2х3 и 3х2 имеют шесть элементов, но сложить их нельзя.

A Верно. Диагональная матрица по определению является симметричной, а значит она не изменяется (инвариантна) при транспонировании.
